```python
#5)
import re
import pandas as pd
from stop_words import get_stop_words
from nltk.corpus import stopwords
import nltk
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 
STOPWORDS = set(stopwords.words('english'))
stop_words = list(get_stop_words('en'))
stopwords = list(set(list(STOPWORDS)) | set(list(stop_words)))
stopwords = [re.sub('[^a-zA-Z0-9]+', '', word) for word in stopwords]
print(len(stopwords))
import os

l_of_w = []
p = os.listdir("/content/preprocessed_cases[cases_29404]")
for folder in p:
    docs = os.listdir("/content/preprocessed_cases[cases_29404]/{}".format((folder)))
    try:
        for file in docs:
            fp = os.p.join("/content/preprocessed_cases[cases_29404]", folder, file)
            
            d = pd.read_csv(fp, sep="\t")
            for col in d.columns:
                d[col] = d[col].str.lower()
                d[col] = d[col].apply(lambda x: re.sub('[^A-Za-z0-9 ]+', '', x))
                d[col] = d[col].apply(lambda x: " ".join(item for item in x.split() if ((item not in stopwords) and (len(item)!= 1))))
                
                for sent in d[col]:
                    l_of_w.extend(sent.split())
       
    except Exception as e:
        print(e)
```

```python
# 1.)
import pandas as pd
url = 'https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab-assignment-04-student-performance.csv'
# reading data from git
df1 = pd.read_csv(url)
for col in df1.columns:
    print("Co:",col)
    print("The no rows:", df1[col].shape[0])
    print("The no u_v column:", df1[col].value_counts().shape[0])
    print(df1[col].value_counts())
    print()
```

```python
# 2.)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.countplot(df1['sh'])
plt.xlabel("S_H")#S_H=Study hours
plt.ylabel("C")#C=Count
plt.show()
```

```python
# 3.)
import random
import seaborn as sns
d_n = df1[['sh','esp']]
for i in d_n.index:
    if d_n['esp'][i] == "Best":
      d_n['esp'][i] = random.uniform(80,100)
    elif d_n['esp'][i] == "Good":
      d_n['esp'][i] = random.uniform(60,79)
    elif d_n['esp'][i] == "Vg":
      d_n['esp'][i] = random.uniform(45,59)
    elif d_n['esp'][i] == "Pass":
      d_n['esp'][i] = random.uniform(30,44)
    elif d_n['esp'][i] == "Fail":
      d_n['esp'][i] = random.uniform(0,29)
sns.factorplot(x='sh',y='esp', kind = 'strip',data=d_n)
```

```python
# 4.)
# Import label encoder 
from sklearn import preprocessing
import seaborn as sns
label_encoder = preprocessing.LabelEncoder()  
def Encoder(df):
    columnsToEncode = list(df.select_dtypes(include=['category','object']))
    le = preprocessing.LabelEncoder()
    # we convert categorical columns data into numerical data
    for feature in columnsToEncode:
        try:
            df[feature] = le.fit_transform(df[feature])
        except:
            print('ER_encod '+feature)
    return df

df12 = Encoder(df1)
# we drop the ms column
df12 = df12.drop('ms', axis = 1)

# we get the correlation data
corr = df12.corr()
sns.set(rc = {'figure.figsize':(20,10)})
sns.heatmap(corr,annot=True)
plt.show()
```
